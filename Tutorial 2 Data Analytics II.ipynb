```python
def hypo(a,b):
    import numpy as nump
    hypotenuse = nump.sqrt(a**2+b**2)
    return hypotenuse
def area(b,h):
    area_of_tri = (b*h)/2.0
    return area_of_tri
def properties(p1,p2):
    hypot = hypo(p1,p2)
    areaT = area(p1,p2)
    return hypot, areaT
```

```python
c = 7.0
d = 8.0

hy, triangle_area = properties(c,d)
```

```python
import numpy as np

def euclidean_distance(x,y):
    e_distance = 0
    e_distance = np.linalg.norm(np.array(x) - np.array(y))
    
    return e_distance
```

```python
point_a = (1,2)
point_b = (2,4)
euc_distance = euclidean_distance(point_a,point_b)
print(euc_distance)
```
